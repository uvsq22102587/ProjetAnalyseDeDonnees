retirer les retours √† la ligne dans le texte du tweet

```python

def no_retourchariot(chiffre):
    """fonction qui retire les retours √† la ligne du texte"""
    global result
    patern = re.compile("\n")
    result = patern.sub(r"",pdTweet["text"][chiffre])
    return result
```

```python

def no_emoticone(chiffre):
        """fonction qui retire tout les √©moticones du tweet"""
        global result
        patern  = re.compile("["
                        u"\U0001F600-\U0001F64F"  # emoticons
                        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                        u"\U0001F680-\U0001F6FF"  # transport & map symbols
                        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                        u"\U00002500-\U00002BEF"  # chinese char
                        u"\U00002702-\U000027B0"
                        u"\U00002702-\U000027B0"
                        u"\U000024C2-\U0001F251"
                        u"\U0001f926-\U0001f937"
                        u"\U00010000-\U0010ffff"
                        u"\u2640-\u2642"
                        u"\u2600-\u2B55"
                        u"\u200d"
                        u"\u23cf"
                        u"\u23e9"
                        u"\u231a"
                        u"\ufe0f"  # dingbats
                        u"\u3030"
                                "]+", flags = re.UNICODE)
        
        result = patern.sub(r"",chiffre)
        return result

```

retirer mention du texte d'un tweet

```python

def no_mention(chiffre):
    """fonction qui retire les mentions du tweet"""
    global result
    patern = re.compile("@\S*")
    result = patern.sub(r"",chiffre)
    return result

```

retirer les hastags d'un tweet

```python

def no_hastags(chiffre):
    """fonction qui retire les hastags du tweet"""
    global result
    patern = re.compile("#\S*")
    result = patern.sub(r"",chiffre)
    return result
```

retirer l'int√©gration d'un URL dans un tweet

```python

def no_URL(chiffre):
    """fonction qui retire l'URL d'un tweet"""
    global result
    patern = re.compile("https://\S*")
    result = patern.sub(r"", chiffre)
    return result
```
